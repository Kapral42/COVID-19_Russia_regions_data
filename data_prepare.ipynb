# Data preporation

```python
file_name = 'covid19-russia-cases.csv'
rus_df = pd.read_csv(file_name, index_col=False)
rus_df.tail(10)
```

```python
# Bug fixes
rus_df['Region/City'] = rus_df['Region/City'].astype('str').str.strip('\u200b') 
print(rus_df['Region/City'])
```

Add region english translation

```python
from googletrans import Translator

def translate(rus):
    eng = translator.translate(rus, src='ru', dest='en')
    return eng.text

translator = Translator()
rus_df['Region/City-Eng'] = rus_df['Region/City'].apply(translate)

# Reorder columns
col_list = rus_df.columns.tolist()
reord_col_list = col_list[:2] + [col_list[-1]] + col_list[2:-1]
rus_df = rus_df[reord_col_list]
rus_df.head()
```
